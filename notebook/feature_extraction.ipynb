```python
import pandas as pd
from tqdm import tqdm
import re
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import json
import time
```

```python
# As the data set is very huge we will only perform our analysis on 1000 playlists
with open("./data/spotify_million_playlist_dataset/data/mpd.slice.7000-7999.json") as f:
    raw = json.load(f)
df = pd.json_normalize(raw["playlists"], record_path='tracks', meta=['name'])
df.shape
```

```python
df["track_uri"] = df["track_uri"].apply(lambda x: re.findall(r'\w+$', x)[0])
df["track_uri"]
```

```python
# We will have to get the features in different subsets or chunks as the dataset is very huge.
# We will have to wait to not cross the rate limit of the Spotify API and get features in batches of 50
#Reference https://stackoverflow.com/questions/74579070/spotipy-sp-track-specifically-takes-too-long-to-run

chunker = lambda y, x: [y[i : i + x] for i in range(0, len(y), x)]
uri_chunks1 = chunker(df["track_uri"].iloc[:10000], 50)
uri_chunks2 = chunker(df["track_uri"].iloc[10000:20000], 50)
uri_chunks3 = chunker(df["track_uri"].iloc[20000:30000], 50)
uri_chunks4 = chunker(df["track_uri"].iloc[30000:40000], 50)
uri_chunks5 = chunker(df["track_uri"].iloc[40000:50000], 50)
uri_chunks6 = chunker(df["track_uri"].iloc[50000:], 50)
cid = "<client_id_from_api_app>"
secret = "<client_secret_from_api_app>"
client_credentials_manager = SpotifyClientCredentials(client_id=cid, client_secret=secret)
sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager)
count = 0
all_features = []
uri_chunks = uri_chunks1
for chunk in uri_chunks:
    num = len(uri_chunks)
    print(f"{count}/{num}")
    features = sp.audio_features(chunk)

    # Get the tracks
    tracks = sp.tracks(chunk)["tracks"]
    artist_ids = []
    for track in tracks:
        artist_ids.append(track["artists"][0]["id"])
    artist_pops = sp.artists(artist_ids)["artists"]
    artists_popularity = []
    artist_genres = []
    for i in range(len(chunk)):
        if i < len(artist_pops[i]):
            artist = artist_pops[i]
            pop = artist["popularity"]
        else:
            pop = 0
        features[i]["artist_pop"] = pop
        features[i]["track_pop"] = tracks[i]["popularity"]
        artist_genres = artist["genres"]
        if artist_genres:
            features[i]["genres"] = " ".join([re.sub(' ','_',i) for i in artist_genres])
        else:
            features[i]["genres"] = "unknown"
    time.sleep(10)
    count += 1
    all_features += features
```
